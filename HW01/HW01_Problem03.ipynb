```python
### 20 New Groups Data set 
# Import data set
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer 
ng_df = fetch_20newsgroups()
```

```python
### Normalize features (TF-IDF) 
vectorizer = TfidfVectorizer() 
ng_df_normalized = vectorizer.fit_transform(ng_df.data) 
# Make to Subset
from sklearn.model_selection import train_test_split  
X_train, X_test = train_test_split(ng_df_normalized, test_size=0.3) 
```

```python
### Compute Similarity matrics 
from sklearn.metrics.pairwise import cosine_similarity, euclidean_distances
similarity_cosine = cosine_similarity(X_train) 
similarity_dot = ng_df_normalized.dot(X_train.T)
similarity_euclidean = euclidean_distances(X_train)
print(similarity_cosine[3:])
print(similarity_dot[3:])
print(similarity_euclidean[3:])
```
