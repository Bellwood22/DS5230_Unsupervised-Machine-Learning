### ***20 NewsGroups Data set***

```python
### Import Data 
# Import data set
from sklearn.datasets import fetch_20newsgroups 
from sklearn.feature_extraction.text import TfidfVectorizer 
import pandas as pd
import numpy as np
df = fetch_20newsgroups() 
### X Normalization 
# vectorize 
vectorizer = TfidfVectorizer() 
categories = ['alt.atheism', 'talk.religion.misc','comp.graphics', 'sci.space']
df = fetch_20newsgroups(subset="train", categories=categories) 
df = vectorizer.fit_transform(df.data) 
print("X Size: ", df.shape) 
```

```python
### Set target
# target names 
categories = ['alt.atheism', 'talk.religion.misc','comp.graphics', 'sci.space']
y = fetch_20newsgroups(subset="train", categories=categories)  
y = y.target
print("target shape: ", y.shape)
```

```python
### Data Split 
from sklearn.model_selection import train_test_split 
X_train, X_test, y_train, y_test = train_test_split(df, y, test_size=0.1) 
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.1)
print("X_train and y_train size: ", X_train.shape, y_train.shape) 
print("X_valid and y_valid size: ", X_valid.shape, y_valid.shape) 
print("X_test and y_test size: ", X_test.shape, y_test.shape) 
```

```python
### Train Model (KNN) 
from sklearn.neighbors import KNeighborsClassifier 
knn = KNeighborsClassifier(n_neighbors=5) 
# Fit
knn.fit(X_train, y_train) 
# predict
y_train_pred = knn.predict(X_train) 
y_valid_pred = knn.predict(X_valid) 
y_test_pred = knn.predict(X_test)
# Accuracy 
from sklearn.metrics import accuracy_score 
print('Train Model accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_train_pred)))
print('Train Model accuracy score: {0:0.4f}'. format(accuracy_score(y_valid, y_valid_pred)))
print('Test Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_test_pred)))
```

```python
print('Training set score: {:.4f}'.format(knn.score(X_train, y_train)))
print('Validation set score: {:.4f}'.format(knn.score(X_valid, y_valid)))
print('Test set score: {:.4f}'.format(knn.score(X_test, y_test)))
```

### ***MNIST Data set***

```python
### Import Data 
from sklearn import datasets
from sklearn.preprocessing import MinMaxScaler, StandardScaler
# Load MNIST dataset
mnist = datasets.load_digits() 
n_samples = len(mnist.images) 
df = mnist.images.reshape((n_samples, -1)) 
# Data Split 
X_train, X_test, y_train, y_test = train_test_split(df, mnist.target, test_size=0.1) 
X_train, X_valid, y_train, y_valid = train_test_split(X_train, y_train, test_size=0.1) 
print("X_train and y_train size: ", X_train.shape, y_train.shape) 
print("X_valid and y_valid size: ", X_valid.shape, y_valid.shape) 
print("X_test and y_test size: ", X_test.shape, y_test.shape) 
```

```python
### Normalization 
scaler = MinMaxScaler() 
X_train = scaler.fit_transform(X_train)
X_valid = scaler.fit_transform(X_valid)
X_test = scaler.fit_transform(X_test)
print("Normalization data size: ", X_train.shape)
```

```python
### Model fitting 
knn = KNeighborsClassifier(n_neighbors=5) 
# Fit
knn.fit(X_train, y_train) 
# predict
y_train_pred = knn.predict(X_train) 
y_valid_pred = knn.predict(X_valid) 
y_test_pred = knn.predict(X_test)
# Accuracy 
from sklearn.metrics import accuracy_score 
print('Train Model accuracy score: {0:0.4f}'. format(accuracy_score(y_train, y_train_pred)))
print('Train Model accuracy score: {0:0.4f}'. format(accuracy_score(y_valid, y_valid_pred)))
print('Test Model accuracy score: {0:0.4f}'. format(accuracy_score(y_test, y_test_pred)))
```

```python
print('Training set score: {:.4f}'.format(knn.score(X_train, y_train)))
print('Validation set score: {:.4f}'.format(knn.score(X_valid, y_valid)))
print('Test set score: {:.4f}'.format(knn.score(X_test, y_test)))
```
