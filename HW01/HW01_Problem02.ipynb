```python
### Drive mount
from google.colab import drive
drive.mount("/content/drive") 
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt 
import webbrowser
```

```python
### Convert dat file to arff
kosarak = "/content/drive/MyDrive/DS5230_Unsupervised Machine Learning/Assignments/HW01/uchoice-Kosarak.txt"
# read kosrak data
with open(kosarak, "r+", encoding="utf-8") as file:
  users = file.readlines() 
  data = set() 
  strings = [] 
  # add num in data
  for user in users:
    for num in list(map(int, user.strip().split(" "))):
      data.add(num) 
  data = range(max(data)) 
  # Generate attribute strings in each file 
  data_strings = [("@attribute news%d {0, 1}" %(num)) for num in data]
  # Set formatted strings 
  strings = ["{%s}" %",".join("%d 1" %(num-1) for num in sorted(list(set((list(map(int, user.strip().split(" ")))))))) for user in users]
  # to_arff
  items = ["@relation kosarak"] + data_strings + ["@data"] + strings 
  # write file 
  path = "/content/drive/MyDrive/DS5230_Unsupervised Machine Learning/Assignments/HW01/result.arff"
  with open(path, "w") as file:
    for line in items:
      file.write(line + "\n")
```
