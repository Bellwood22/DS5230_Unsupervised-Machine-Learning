```python
### Import Library & Data 
# Library
from google.colab import drive
drive.mount("/content/drive") 
import numpy as np
import pandas as pd
import random
# df
df = "/content/drive/MyDrive/DS5230_Unsupervised Machine Learning/Assignments/HW02/coin_flips_outcome.txt"
df = pd.read_csv(df, sep=" ", header=None)  
df = np.array(df)
```

```python
### Compute probabilities
# Calculate probabilities for each sessions
probas = [] 
n = 20 
for i in range(len(df)): 
  head_ = 0
  proba = 0
  for j in range(len(df[0])): 
    if df[i][j] == 1:
      head_ += 1
      proba = head_ / n 
  probas.append(proba) 
probas = np.array(probas)
# Get the np of probas 
np_proba = 1
for i in range(len(df)):
  np_proba *= probas[i]
coin1_proba = np_proba * 50 
coin2_proba = np_proba * 50 * 1/2
coin3_proba = np_proba * 50 * 1/3
np_probas = [coin1_proba,coin2_proba,coin3_proba]
np_probas
```

```python
### Calculate bias with probas 
orgin_proba = 1/2 
bias = []
for i in range(len(df)):
  bia = probas[i] - orgin_proba
  bias.append(bia)
bias = np.array(bias)
np_bias = 1
for i in range(len(df)):
  np_bias *= bias[i]
coin1_np_bias = np_bias * 50 
coin2_np_bias = np_bias * 50 * 1/2
coin3_np_bias = np_bias * 50 * 1/3
np_bia = [coin1_proba,coin2_proba,coin3_proba]
np_bia
```
