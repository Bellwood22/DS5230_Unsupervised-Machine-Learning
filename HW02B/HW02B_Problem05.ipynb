```python
### Import library & data set 
# library
from google.colab import drive
drive.mount("/content/drive") 
import numpy as np
import pandas as pd
import random
from sklearn.cluster import DBSCAN
import matplotlib.pyplot as plt
import seaborn as sns
# df
df = "/content/drive/MyDrive/DS5230_Unsupervised Machine Learning/Assignments/HW02B/dbscan.csv"
df = pd.read_csv(df)  
df = pd.DataFrame(df)
df.head()
```

```python
### DBSCAN - sklearn
# creat dbscan
X = df[["x","y"]] 
X = np.array(X) 
dbscan = DBSCAN(eps=7.5, min_samples=3).fit(X) 
df["cluster"] = dbscan.labels_ 
# plt result
sns.scatterplot(data=df, x="x", y="y", hue="cluster") 
```
